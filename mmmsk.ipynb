```python
categorical_columns  = data.dtypes[data.dtypes == 'object'].index
```

## Count , Proportion , Mode 

```python
for col in categorical_columns:
    print('Column :- ' , col)
    print(data[col].value_counts())
    print(f'Proportion of Each Category    ' , data[col].value_counts()/data[col].count() )
    # Mode is defined as the category having highest proportion 
    print(f'Mode of the column is :  ' , data[col].value_counts().index[0] )
```

## Mean Median Mode 

```python
numerical_columns = data.dtypes[data.dtypes != 'object'].index
```

```python
for col in numerical_columns:
    print('Column : ', col )
    print('Mean of the Column : ' , data[col].mean())
    print('Median of the column is : ',data[col].median())
```

## Range , IQR , Variance , Standard Deviation , Coefficient of Variation , Z-Score , Skewness 

```python
for col in numerical_columns:
    print('Column : ', col)
    print('Range of the Column is  ' , data[col].max()-data[col].min())
    print('Inter Quantile Range of thge Column is  : ', data[col].quantile(0.75)-data[col].quantile(0.25))
    print('Variances of the Column is : ' , data[col].std()**2)
    print('Standard Deviation of thge Column is : ', data[col].std())
    print('Skewness of the Column is : ' , np.where(3*(data[col].mean()-data[col].median())/data[col].std()<0 , 'Left Skewed' , 'Right Skewed'))
    print('Coefficient of Variation of the Column is : ', data[col].std()/data[col].mean())
```

#### Skewness is defined as follows : 

    Left Skewed (or) Negitively Skewed :  if mean < median < mode 

    Symmetric : Mean = Median = Mode

    Right Skewed (or) Positively Skewed : if mean > median > mode 

##### From above we came to know that skewness is dependent on the relative positions of mean and median if mean < median {Left Skewed} or mean > Medoan {Right Skewed}

#### Calculation of Skewness

Criteria - 1 : 
        
        Pearson's Coefficient of Skewness : 
                            sk = 3(mean - median)/standardDeviation
Criteria - 2 :      
  
        Bowley's Coefficient of Skewness :
                            sk = ((Q3-Q2)-(Q2-Q1))/((Q3-Q2)+(Q2-Q1))
